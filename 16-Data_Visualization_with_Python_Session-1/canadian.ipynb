```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("canadian_immegration_data.csv",index_col="Country")
data
```

```python
data.info()
```

```python
data.describe(include="O")
```

```python
de=data[['1991','1993','1995','1997','1999']]
de.head()
```

```python
# max unmber of immigrants in [1991 : 1999]
plt.plot(de.max(),"-b")
plt.title("immigration")
plt.xlabel("Years")
plt.ylabel("number of immigrants")
plt.show()
```

```python
iceland_data = data.loc["Iceland"]
# iceland_data
values = iceland_data[["1991", "1992","1993","1994", "1995","1996","1997", "1998","1999"]]
values
labels = ["1991", "1992","1993","1994", "1995","1996","1997", "1998","1999"]
labels
```

```python
# number of iceland immigration to canada from [1991:1999]
plt.bar(labels,values)
plt.title("iceland immigration to canada")
plt.xlabel("years")
plt.ylabel("number of immigrants")
plt.show()
```

```python
egy_data = data.loc["Egypt"]
valuesE = egy_data[["1991", "1992","1993","1994", "1995","1996","1997", "1998","1999","2000"]]
labelsE = ["1991", "1992","1993","1994", "1995","1996","1997", "1998","1999","2000"]
lyb_data = data.loc["Libya"]
valuesl = lyb_data[["1991", "1992","1993","1994", "1995","1996","1997", "1998","1999","2000"]]
labelsl = ["1991", "1992","1993","1994", "1995","1996","1997", "1998","1999","2000"]
valuesl
```

```python
# the difference  between egyption and Libyian immigration to canada
plt.plot(labelsE,valuesE)
plt.plot(labelsl,valuesl)
plt.title("egyption and Libyian immigration to canada")
plt.xlabel("years")
plt.ylabel("number of immigrants")
plt.show()
```

```python
Ind_data = data.loc["India"]
valuesI = Ind_data[["1991", "1992","1993","1994", "1995","1996","1997", "1998","1999","2000"]]
labels = ["1991", "1992","1993","1994", "1995","1996","1997", "1998","1999","2000"]
Ger_data = data.loc["Germany"]
valuesG = Ger_data[["1991", "1992","1993","1994", "1995","1996","1997", "1998","1999","2000"]]
Far_data = data.loc["France"]
valuesF = Far_data[["1991", "1992","1993","1994", "1995","1996","1997", "1998","1999","2000"]]
```

```python
plt.scatter(labels,valuesI,color="red")
plt.scatter(labels,valuesG,color="green")
plt.scatter(labels,valuesF,color="blue")
plt.title("India , Germany and France immigration to canada")
plt.xlabel("years")
plt.ylabel("number of immigrants")
plt.legend(labels=["India","Germany","France"], title="Categories", loc="upper right")
plt.show()
```

```python

plt.bar(["egypt"],data.loc["Egypt"][["Total"]])
plt.bar(["Brazil"],data.loc["Brazil"][["Total"]])
plt.bar(["Dominica"],data.loc["Dominica"][["Total"]])
plt.bar(["France"],data.loc["France"][["Total"]])
plt.bar(["Germany"],data.loc["Germany"][["Total"]])
plt.bar(["Iraq"],data.loc["Iraq"][["Total"]])
plt.bar(["Pakistan"],data.loc["Pakistan"][["Total"]])
plt.bar(["Saudi Arabia"],data.loc["Saudi Arabia"][["Total"]])
plt.show()
```

```python
lable=list(map(str,range(1980,2014)))
iceland_data=data.loc["Iceland",lable]
iceland_data
```

```python
plt.bar(lable,iceland_data,color="magenta")
plt.title("iceland immigration to canada")
plt.xticks(rotation=90)
plt.xlabel("year")
plt.ylabel("number of immigration")
plt.grid()
plt.show()
```

```python
df=data.groupby("Continent",axis=0).sum()
explode = [0.1, 0, 0, 0,0,0] 
labels=["Africa","Asia","Europe","Latin America","Northern America","Oceania"]
plt.pie(df["Total"],
        labels=labels,             
        autopct='%1.1f%%', 
        explode=explode,
        shadow=True,           # Explode the first slice
 )
plt.legend(labels, title="Categories", loc="upper left")

plt.show()
```

```python
year=list(map(str,range(1980,2013)))
df=data.loc["Japan"][year]
plt.boxplot(df)
plt.xlabel("Japan")
plt.ylabel("number of immigration")
plt.title(" number of japanis immigration from 1980 to 2013")
plt.show()
```
