## Step 4: Correlation Heatmap  
We will use a **heatmap** to visualize the correlation between numerical variables.  
This helps us identify which variables have strong positive or negative correlations.

```python
# Create a heatmap of correlations between numerical variables
# Compute the correlation matrix
correlation_matrix = data.corr(numeric_only=True)  # Compute correlation for numeric columns

# Convert to long format for Plotly visualization
corr_long = correlation_matrix.reset_index().melt(id_vars="index")

# Rename columns for better readability
corr_long.columns = ["Variable 1", "Variable 2", "Correlation"]

# Create a heatmap using Plotly
fig =sns.heatmap(
    correlation_matrix,
 
     annot=True, cmap="coolwarm", fmt=".2f", cbar=True, linewidths=0.5

)
fig.show()
# Customize the plot for better readability
# plt.title('Correlation Heatmap')  
# plt.show()
```

## Step 5: Bar Plot with an Extra Variable (Hue)  
In this plot, we will analyze **MonthlyIncome** for each **Department**, while adding **Attrition** as a third dimension (hue).  
This allows us to compare income levels across departments and see how attrition status affects them.

```python
# Create a bar plot with Attrition as the hue (extra variable)
sns.barplot(data=data,
            x="Department",
            y="MonthlyIncome",
            hue="Department")

# Customize the plot with title and axis labels
plt.title('Monthly Income by Department and Attrition')
plt.xlabel('Department')
plt.ylabel('Monthly Income')

# Rotate x-axis labels for better readability

```
