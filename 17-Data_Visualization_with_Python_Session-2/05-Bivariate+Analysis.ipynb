## Step 1: Import Necessary Libraries  
We begin by importing the essential libraries:  
- **pandas** for data manipulation  
- **matplotlib.pyplot** for plotting  
- **seaborn** for advanced data visualization

```python
# Importing pandas for data handling, matplotlib for plotting, and seaborn for creating advanced plots
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 2: Load the Dataset  
Load the **employee dataset** using pandas' `read_csv()` function and check its shape to understand its size.

```python
# Read the dataset
data=pd.read_csv('datasets\employee.csv')

# Check the number of rows and columns in the dataset
# Returns the number of rows and columns as a tuple
print(data.shape)
```

## Step 3: Explore the Dataset  
It's always a good idea to preview the first few rows of your dataset to understand its structure and content.

```python
# Display the first five rows of the dataset
data.head()
```

## Step 4: Categorical vs Continuous Variables  
We will use a **bar chart** to visualize the **average MonthlyIncome** for each **Department**.  
This helps in understanding how income differs across different departments.

```python
# Create a bar plot to show the average MonthlyIncome for each Department
# plt.bar(data=data,
#             x="MonthlyIncome".mean(),
#             y="Department"
#             )
avg_income = data.groupby("Department")["MonthlyIncome"].mean()
plt.bar(x=avg_income.index,
        height=avg_income.values,
        color="skyblue")
# Customize the plot with title and axis labels
plt.title('Average Monthly Income by Department')
plt.xlabel('Department')
plt.ylabel('Average Monthly Income')

# Rotate x-axis labels for better readability
# plt.xticks(rotation=45)
```

## Step 5: Continuous vs Continuous Variables  
Next, we will use a **scatter plot** to visualize the relationship between **Age** and **MonthlyIncome**.  
This helps in identifying patterns or correlations between these two continuous variables.

```python
# Create a scatter plot to show the relationship between Age and MonthlyIncome
plt.scatter(data=data,
            x="Age",
            y="MonthlyIncome"
            )

# Customize the plot with title and axis labels
plt.title('Relationship between Age and Monthly Income')
plt.xlabel('Age')
plt.ylabel('Monthly Income')

# Add a grid to the plot for easier interpretation
plt.grid()
```
